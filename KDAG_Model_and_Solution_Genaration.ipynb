# one HCP at a time

```python
final_df = df.groupby('Physician_ID').mean()


calls = 1
emails = 9
samples = 159

options = []

for i in range(1,calls+1):
  temp=[i,0,0]
  options.append(temp)
for i in range(1,emails+1):
  temp=[0,i,0]
  options.append(temp)
for i in range(1,samples+1):
  temp=[0,0,i]
  options.append(temp)

n = len(options)



```

```python
def predict_score(X, model):
    prediction_brandRx = model.predict(X)
    return prediction_brandRx # it is a 10000*1 vector

data_cols = final_df.columns
```

```python
data_cols
```

```python
all_scores = np.zeros((10000,n))

for i in range(10000):
  for j in range(n):
    predictor_vals = options[j]
    data_row = final_df.iloc[i]
    # print("1")
    # print(data_row)
    data_row = pd.DataFrame(data_row).T
    data_row.reset_index(inplace=True)
    data_row.drop(['Brand_Rx', 'index'], axis=1, inplace=True)
    # print("11")
    # print(type(data_row))
    # print(data_row)
    data_row['Sales_Rep_Calls'] = predictor_vals[0]
    data_row['Emails_Delivered'] = predictor_vals[1]
    data_row['Samples_Dropped'] = predictor_vals[2]
    # print("2")
    # print(type(data_row))
    # print(data_row)
    # # data_row = pd.DataFrame(data_row.values).T
    # # data_row = data_row.reset_index()
    # data_row = pd.DataFrame(data_row.values, columns=data_cols)
    # data_row.drop(['Brand_Rx'], axis=1, inplace=True)
    # print("3")
    # print(type(data_row))
    # print(data_row)
    all_scores[i][j] = predict_score(data_row,XGBoost)
```

```python
pd.DataFrame(all_scores).to_csv('/content/scores.csv')
```

```python
answer = np.zeros(10000)
tot_samples = 0

# 0 - calls-1 : calls 0
# calls - calls+emails-1 : emails 2
# calls+emails - end : samples 1


def possible(idx, id):
  calls_, emails_, samples_ = options[idx]
  curr = df_hcp_sum.iloc[id]
  seg = curr['Physician_Segment']/57 # number of weeks, since it is the sum
  rec_calls = curr['Sales_Rep_Calls']
  if tot_samples+samples_ > 2500:
    return False
  if seg == 3 and calls_+rec_calls > 48:
    return False
  if seg == 2 and calls_+rec_calls > 24:
    return False
  if seg == 1 and calls_+rec_calls > 12:
    return False
  return True

brand_Rx = np.zeros(10000)
for i in range(10000):
  flag = 1
  sorted_idx = (-all_scores.iloc[i]).argsort()
  j = 0 # iterator over sorted_idx which stores the indices of the sorted list
  idx = sorted_idx[j]
  while(not possible(idx, i)):
    j+=1
    if j >= n:
      answer[i] = 2
      brand_Rx[i] = all_scores.iloc[i, 1:10].max()
      flag = 0
      break
    idx = sorted_idx[j]
  calls_, emails_, samples_ = options[idx]
  if flag:
    continue
  brand_Rx[i] = all_scores.iloc[i,idx] 
  if calls_:
    answer[i] = 0
  elif emails_:
    answer[i] = 2
  elif samples_:
    answer[i] = 1
```

```python
pd.DataFrame(answer).to_csv('/content/answer.csv')
```
